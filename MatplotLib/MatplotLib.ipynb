```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('X-Y Plot')
plt.show()
```

# Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])   #[Left,Bottom,Width,Height] and range from 0 to 1
axes2=fig.add_axes([0.2,0.4,0.5,0.7])
axes1.plot(x,y)
axes2.plot(x,y)
```

 Creating SubPlot

```python
fig,axes=plt.subplots(nrows=1,ncols=2)  #This divides the plot into i row and 2 colunms
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)  
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
axes  #It is a list of the matplotlib axes
```

```python
#Second method for plotting 

fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure Size , Aspect Ratio  and DPI

```python
fig=plt.figure(figsize=(7,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,4))
axes[0].plot(x,y)
axes[1].plot(y,x,label='x,y')   #This creates a note indicating the seperate plots
axes[1].plot(x,y,label='y,x')
axes[1].legend(loc=0)          #Loc =0 automatically determines the best location for the labels on the plot
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'red',linewidth=5,alpha=0.5,linestyle='steps')   
#Customize the color of the plot. We can also use the RGB Hex Code
#linewidth : Determines the width of the line
#aplha : Determines the transparency of the plotted line
#linestyle: THe pattern of the line
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'red',linewidth=5,alpha=0.5,linestyle='--')   
#Customize the color of the plot. We can also use the RGB Hex Code
#linewidth : Determines the width of the line
#aplha : Determines the transparency of the plotted line
#linestyle: THe pattern of the line 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'red',linewidth=5,alpha=0.5,linestyle=':')   
#Customize the color of the plot. We can also use the RGB Hex Code
#linewidth or lw : Determines the width of the line
#aplha : Determines the transparency of the plotted line
#linestyle or ls : THe pattern of the line 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'red',linewidth=2,alpha=0.6,linestyle='--',marker='o',markersize=5,markerfacecolor='yellow',markeredgewidth=15)   
#Customize the color of the plot. We can also use the RGB Hex Code
#linewidth : Determines the width of the line
#aplha : Determines the transparency of the plotted line
#linestyle: THe pattern of the line 
#marker : Marks the coordinate points i.e. the x and y coordinate of the graph
#markersize : Gives the size of marker point
```

# Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([01.,0.1,0.8,0.8])
ax.plot(x,y)
ax.set_xlim([0,1])   #Initially the range of plot was from 0-25 but 'xlim' helps to limit the plot range
ax.set_ylim([0,1]) 
```
