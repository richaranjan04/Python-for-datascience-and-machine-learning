```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
 sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#Segregated male and female
#sactter_kws : controlles the size of the shapes 'o' and 'v'
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':50})  
```

```python
#This creates two idfferent columns of data according to the gender
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')        
```
