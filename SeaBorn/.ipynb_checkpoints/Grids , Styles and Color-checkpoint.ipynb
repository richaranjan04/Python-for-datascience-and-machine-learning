```python
import seaborn as sns
%matplotlib inline
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map_upper(sns.scatterplot)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
g= sns.FacetGrid(data=tips,row='smoker',col='time')
g.map(sns.distplot,'total_bill')
```

# Styles and color

```python
sns.set_style('whitegrid')  #Or white,ticks etc
sns.countplot(x='sex',data=tips)
```

```python
#Chnage size of plot
sns.set_context('poster',font_scale=2)    # or Paper,notebook etc. and font-scale controlles the font size
sns.countplot(x='sex',data=tips)
```
