```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)

#Plots according to the average of the parameters of comparison
#Hence from graph mean of male total_bill > female's total_bill
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)

#Plots according to the standard deviation of the total_bill vs gender
```

```python
sns.countplot(x='sex',data=tips)   #Gives comparison of count  of entities in the dataset
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)  #combination of voilin plot and strip plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black') 

#Gives combination of violin plot and swarm plot on top of voilen plot
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
